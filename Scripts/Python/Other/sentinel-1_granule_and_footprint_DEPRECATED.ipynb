```python
# func to convert the multipolygon string for each granule to float
def foot_to_float(footprint_str):
    coordinate_pairs_str = footprint_str.split("(((")[1].split(")))")[0].split(",")
    coordinate_pairs = []
    for pair_str in coordinate_pairs_str:
        temp = []
        lat_str, lon_str = pair_str.strip().split()
        # temp = [float(lat_str), float(lon_str)]
        temp = [float(lon_str), float(lat_str)]
        coordinate_pairs.append(
            temp
        )  # Swap lat and lon to follow (lon, lat) convention
    pairs_within_pairs = [coordinate_pairs]
    return pairs_within_pairs
```
