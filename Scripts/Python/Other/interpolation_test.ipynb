```python
import numpy as np
from scipy.interpolate import interp1d

# Generate some example data
x = np.linspace(0, 10, 10)  # X coordinates of points
y = np.sin(x)                # Y coordinates of points

# Define the interpolation function
interp_func = interp1d(x, y, kind='linear')

# Generate new points for interpolation
x_new = np.linspace(0, 10, 100)  # New X coordinates for interpolation

# Interpolate the new points
y_new = interp_func(x_new)

# Plot the original points and the interpolated curve
import matplotlib.pyplot as plt
plt.plot(x, y, 'o', label='Original Points')
plt.plot(x_new, y_new, '-', label='Interpolated Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Interpolation of Point Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt

# Assuming you have your raster data with latitude, longitude, and values
# For this example, let's generate some random data
np.random.seed(0)
n_points = 100
lats = np.random.uniform(-90, 90, n_points)
lons = np.random.uniform(-180, 180, n_points)
values = np.random.rand(n_points)

# Define the grid to interpolate onto
lat_grid, lon_grid = np.meshgrid(np.linspace(-90, 90, 100), np.linspace(-180, 180, 200))

# Interpolate the values onto the grid
interp_values = griddata((lats, lons), values, (lat_grid, lon_grid), method='linear')

# Plot the original points and the interpolated grid
plt.figure(figsize=(10, 5))

# Plot original points
plt.scatter(lons, lats, c=values, cmap='viridis', label='Original Points')

# Plot interpolated grid
plt.imshow(interp_values, extent=(-180, 180, -90, 90), origin='lower', cmap='viridis', alpha=0.5, aspect='auto')
plt.colorbar(label='Interpolated Values')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Interpolation of 2D Raster Data')
plt.grid(True)
plt.legend()
plt.show()
```
